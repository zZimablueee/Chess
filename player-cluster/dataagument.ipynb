```python
#negetive增强
augmented_rows = []

# 定义需要排除计算的列（除了题目要求的uid和color，player也要先排除再加回来）
cols_to_exclude = ['uid', 'color', 'player']
cols_to_convert = [c for c in Dynamic_0_all.columns if c not in cols_to_exclude]
for col in cols_to_convert:
    Dynamic_0_all[col] = pd.to_numeric(Dynamic_0_all[col], errors='coerce')

# 遍历每个目标选手
for player_name in Dynamic_negetive_names:
    # 1. 在大表中找到该选手的所有数据
    player_data = Dynamic_0_all[Dynamic_0_all['player'] == player_name]
    
    # 安全检查：如果该选手数据不足10条，无法进行不重复采样
    if len(player_data) < 10:
        print(f"跳过选手: {player_name} (数据行数不足10行)")
        continue

    # 2. 重复10次操作
    for _ in range(10):
        # 3. 随机选取10条数据 (replace=False 表示不重复抽样，即取10个不同的行)
        sample = player_data.sample(n=10, replace=False)
        
        # 4. 剔除不需要求平均的列 (只保留数值列进行计算)
        # errors='ignore' 防止列不存在时报错
        numeric_sample = sample.drop(columns=cols_to_exclude, errors='ignore')
        
        # 5. 每列求平均，得到一条新数据
        # numeric_only=True 确保只对数值类型的列求平均，防止报错
        avg_row = numeric_sample.mean()
        
        # 6. 把选手名字加回去
        avg_row['player'] = player_name
        
        # 收集结果
        augmented_rows.append(avg_row)

# 7. 生成最终的 Dynamic_0_agument 表
Dynamic_0_agument = pd.DataFrame(augmented_rows)

# 调整列顺序：通常把 'player' 放到第一列看起来更直观 (可选)
cols = ['player'] + [c for c in Dynamic_0_agument.columns if c != 'player']
Dynamic_0_agument = Dynamic_0_agument[cols]
Dynamic_0_agument.columns = [col[:-4] if col.endswith("Mean") else col for col in Dynamic_0_agument.columns]
Dynamic_0_agument = Dynamic_0_agument.rename(columns={"player": "Name"})
# 查看结果
print(f"新表生成完毕，共 {len(Dynamic_0_agument)} 行")
print(Dynamic_0_agument.shape)
```
